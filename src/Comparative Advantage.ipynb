# Comparative Advantage

This notebook will do some calculations with comparative advantage concepts

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
columns = ['Meat', 'Potatoes']
data = np.array([[1, 3], [3, 6]])
index = ['Farmer', 'Rancher']

df = pd.DataFrame(data = data, columns = columns, index = index)
df
```

## Production possibilities frontiers

### Rancher

```python
row = 'Rancher'
rancher_max_meat = df.loc[row, 'Meat']
rancher_max_potatoes = df.loc[row, 'Potatoes']
rancher_mid_meat = rancher_max_meat/2
rancher_mid_potatoes = rancher_max_potatoes/2
rancher_trade_meat = 13/8
rancher_trade_potatoes = 27/8

plt.plot([rancher_max_meat, 0], [0, rancher_max_potatoes], label = "Rancher's efficient line")
plt.title("Rancher's production possibilities")
plt.xlabel('Meat')
plt.ylabel('Potatoes')
plt.plot(rancher_mid_meat, rancher_mid_potatoes, 'ro', label = "No trade output")
plt.plot(rancher_trade_meat, rancher_trade_potatoes, 'bo', label = 'Production with trade')
plt.legend(loc = 'lower left')
plt.grid()
```

### Farmer

```python
row = 'Farmer'
farmer_max_meat = df.loc[row, 'Meat']
farmer_max_potatoes = df.loc[row, 'Potatoes']
farmer_mid_meat = farmer_max_meat/2
farmer_mid_potatoes = farmer_max_potatoes/2
farmer_trade_meat = 5/8
farmer_trade_potatoes = 17/8

plt.plot([farmer_max_meat, 0], [0, farmer_max_potatoes], label = "Farmer's efficient line")
plt.title("Farmer's production possibilities")
plt.xlabel('Meat')
plt.ylabel('Potatoes')
plt.plot(farmer_mid_meat,farmer_mid_potatoes, 'ro', label = 'No trade output')
plt.grid()
plt.plot(farmer_trade_meat, farmer_trade_potatoes, 'bo', label = 'Production with trade')
plt.legend(loc = 'lower left')
```

```python
comparison_columns = ['Without Trade', 'With Trade']
gains = pd.DataFrame(
    data = [
        [farmer_mid_meat, farmer_trade_meat],
        [farmer_mid_potatoes, farmer_trade_potatoes],
        [rancher_mid_meat, rancher_trade_meat],
        [rancher_mid_potatoes, rancher_trade_potatoes]
    ], 
    columns = comparison_columns, 
    index = ['Farmer Meat', 'Farmer Potatoes', 'Ranger Meat', 'Ranger Potatoes']
)

gains['Gains'] = gains['With Trade'] - gains['Without Trade']
gains['% Gains'] = gains['Gains'] / gains['Without Trade'] * 100

gains
```
